```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read testing error data 
test_errors_df=pd.read_hdf('test_errors.h5',header=0)
# Filter the subset for station 'GRAS'
subset_indices = test_errors_df.index.get_level_values(0) == 'GRAS'
subset = test_errors_df[subset_indices]
```

```python
plt.figure(figsize=(10, 5))  

# Plot the first column (y_test)
plt.plot(subset.index.get_level_values(1).tolist(),subset['y_test'].tolist(), label='y_test', alpha=0.7)

# Plot the second column (y_test_pred)
plt.plot(subset.index.get_level_values(1).tolist(),subset['y_test_pred'].tolist(), label='y_test_pred', alpha=0.7)

# Add a legend to the plot
plt.legend()

# Add a title and axis labels
plt.title('reference value and prediction value for station GRAS')
plt.xlabel('Date')
plt.ylabel('Values [$kg/m^2$]')

# Add horizontal lines at major y-ticks for visual aid
ax = plt.gca()


ticks = ax.get_yticks()

# 
for tick in ticks:
    plt.axhline(y=tick, color='gray', linestyle='--', alpha=0.7)  # 0.7


plt.show()
```

```python
# Plot for station "MSGR"
train_errors_df=pd.read_hdf('train_errors.h5',header=0)
subset_indices = train_errors_df.index.get_level_values(0) == 'MSGR'
subset = train_errors_df[subset_indices]

plt.figure(figsize=(10, 5))  


plt.plot(subset.index.get_level_values(1).tolist(),subset['y_train'].tolist(), label='y_train', alpha=0.7)

plt.plot(subset.index.get_level_values(1).tolist(),subset['y_train_pred'].tolist(), label='y_train_pred', alpha=0.7)


plt.legend()


plt.title('reference value and prediction value for station MSGR')
plt.xlabel('Date')
plt.ylabel('Values [$kg/m^2$]')


ax = plt.gca()


ticks = ax.get_yticks()


for tick in ticks:
    plt.axhline(y=tick, color='gray', linestyle='--', alpha=0.7)  


plt.show()
```
