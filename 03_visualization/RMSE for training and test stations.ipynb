```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mse_df1 = pd.read_csv('mse_training_station.csv')
mse_df2 = pd.read_csv('mse_test_station.csv')
```

```python
# Calculate the square root of MSE values
rmse_values1 = np.sqrt(mse_df1['MSE'])
rmse_values2 = np.sqrt(mse_df2['MSE'])

rmse_values1 = np.where(rmse_values1 > 5, 5, rmse_values1)
rmse_values2 = np.where(rmse_values2 > 5, 5, rmse_values2)

# Calculate the minimum and maximum values for bins
min_value = min(rmse_values1.min(), rmse_values2.min())
max_value = max(rmse_values1.max(), rmse_values2.max())

# Create bins array with 20 bins plus boundaries
bins = np.linspace(min_value, max_value, num=21)  

# Plot histograms
plt.hist(mse_values1, bins=bins, alpha=0.5, label='training station', edgecolor='white')
plt.hist(mse_values2, bins=bins, alpha=0.5, label='test station', edgecolor='white')

# Add legend, title, and labels
plt.legend(loc='upper right')
plt.title('Distribution of RMSE')
plt.xlabel('Values [$kg/m^2$]')
plt.ylabel('# stations')

# Add horizontal lines at major y-ticks
ax = plt.gca()
ticks = ax.get_yticks()

for tick in ticks:
    plt.axhline(y=tick, color='gray', linestyle='--', alpha=0.7)  


plt.show()
```
